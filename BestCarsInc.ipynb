<h1>Cleaning data</h1>

Remove redundant and sparse features
* Remove redundant or unusable data: drop the 'URL,' 'image_url,' and 'city_url' features.

```python
#We are going to proceed to eliminate the columns mentioned above
df.drop(columns=['url','image_url','region_url'], axis= 1, inplace= True)
```

```python
#We print the df to make sure that the columns have been correctly deleted.
df.head()
```

```python
#Let's see which columns have more than 40% NaN in their data.
df_percentage = df.isnull().mean()
df_percentage
```

```python
#Now we know which columns to remove
threshold = 0.4
columns_to_drop = df_percentage[df_percentage > threshold].index.tolist()
columns_to_drop
```

```python
#We proceed to eliminate these columns
df = df.drop(columns= columns_to_drop)
df.head()
```

<h2>Remove outliers</h2>
We detected an outlier problem with ‘price,’ ‘year,’ and ‘odometer.’ 


Set a reasonable range for the three features. Explain your decision.
Remove the data point (rows) that doesn’t comply with your defined ranges.
Remove any data point with null values.
What’s the resulting number of rows? What percentage of rows did we remove due to the cleaning and removing of outliers? Is the dataset still usable? Why?

```python
#First we make the decision to remove the NaN values from the columns to be used.
df['year'].dropna(inplace=True)
df['odometer'].dropna(inplace=True)
```

```python
# Let's calculate the IQR of the 'price' column.
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Then we calculate the lower and upper limits for the outliers.
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Now we can identify the outliers in the 'price' column.
outliers_df = df[(df['price'] < lower_limit) | (df['price'] > upper_limit)]

df.drop(index = outliers_df.index, inplace = True)

print(df)
```

```python
#We proceed to identify and eliminate the outliers in the oddometer column.
outliers_df = df[(df['odometer'] < lower_limit) | (df['odometer'] > upper_limit)]
df.drop(index = outliers_df.index, inplace = True)


print(df)
```

<h2>Feature Relations</h2>

```python
df_numeric = df_filtrado.select_dtypes(include=['number'])
correlation_matrix = df_numeric.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
plt.scatter(df_filtrado['price'], df_filtrado['odometer'], alpha=0.5) 
plt.xlabel('Price')
plt.ylabel('Odometer')
plt.title('Relationship between Price and Odometer')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df_filtrado, vars=['price', 'year', 'odometer'])
plt.suptitle("Pairplot of Numerical Variables", y=1.02)
plt.show()
```

<li>Plot a histogram of the feature ‘odometer.’ Use 50 bins for the plot. You’ll see a frequency maximum for a specific odometer value range. Explain it.</li>

```python
from matplotlib.ticker import ScalarFormatter
# We set the number of intervals for the histogram
num_bins = 50

# We create the histogram
plt.figure(figsize=(10, 6))
plt.hist(df_filtrado['odometer'], bins=num_bins, edgecolor='k', alpha=0.75)
plt.xlabel('Odometer')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True)

#We change the format of the y-axis to show real numbers.
formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(False)
formatter.set_powerlimits((-1, 1))
plt.gca().xaxis.set_major_formatter(formatter)

plt.show()
```
